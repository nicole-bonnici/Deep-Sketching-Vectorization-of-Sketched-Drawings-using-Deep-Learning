## Second Step of data augmentation: Resizing the sketches and targets


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline 
```

```python
#function to rescale and save target
def downscale (image, factor, k):
    
    new_dim_1 =(int((image.shape[1]/factor)*6),(int(image.shape[0]/factor)*6))
    new_img_1 = cv2.resize(src=image, dsize=new_dim_1, interpolation=cv2.INTER_AREA)
    
    sfile = 'R_Target/{}.jpg'.format(k,factor)
    cv2.imwrite(sfile, (new_img_1))
    
    
```

```python
#function to rescale and save inpur
def downscale_s (image, factor, k):

    new_dim_1 =(int((image.shape[1]/factor)*6),(int(image.shape[0]/factor)*6))
    new_img_1 = cv2.resize(src=image, dsize=new_dim_1, interpolation=cv2.INTER_AREA)
    
    sfile = 'R_Input/{}.jpg'.format(k)
    cv2.imwrite(sfile, (new_img_1))
```
